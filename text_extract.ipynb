```python
import PyPDF2
import pandas as pd
import numpy as np
import re
 
#create file object variable
#opening method will be rb
pdffileobj=open('./data/F00011055-WVS_7_Codebook_Variables_report.pdf','rb')
 
#create reader variable that will read the pdffileobj
pdfreader=PyPDF2.PdfFileReader(pdffileobj)
 
def get_answer_dict(question_nr):
    this_question =  'Q'+str(question_nr)
    next_question = 'Q'+str(question_nr+1)
    
    #print(text.split(this_question)[1].split(next_question)[0].split('\n \n'))
    q_a_list = list(map(lambda a : a.replace('\n','').replace('-.-','.-').replace('.-',''),text.split(this_question)[1].split(next_question)[0].split('\n \n')))

    #a_list = list(filter(lambda a:a.strip() != '',q_a_list))[3:]
    theme = list(filter(lambda a:a.strip() != '',q_a_list))[0].split(':')[0]
    try:
        sub_theme = list(filter(lambda a:a.strip() != '',q_a_list))[0].split(':')[1].strip()
    except BaseException:
        sub_theme = theme
    #print("SUB THEME WAS: " + sub_theme)
    question = list(filter(lambda a:a.strip() != '',q_a_list))[1] + ' ' + list(filter(lambda a:a.strip() != '',q_a_list))[2]

    #a_list = text.split(this_question)[1].split(next_question)[0].replace('\n','').split(sub_theme)[2].strip().split()
    a_list = re.findall("\d.+",text.split(this_question)[1].split(next_question)[0].replace('\n',''))[0].strip().split('  ')[0].split()

    current_number = 0
    current_question = ''
    previous_was_question = False
    a_dict = {}

    for i in range(len(a_list)):
        
        ## If a number
        if '.-' in a_list[i]:
            
            ## If previous token was a not a number, that last question is now finished
            if previous_was_question:
                #print(current_number)
                #print(current_question.strip())
                a_dict[current_number]=current_question.strip()
                current_question = ''
            
            current_number = a_list[i].replace('-.-','.-').replace('.-','')
        else:
            previous_was_question = True
            current_question += ' ' + a_list[i]
            
            # Add last question
            if i+1 == len(a_list):
                #print(current_number)
                #print(current_question.strip())
                a_dict[current_number]=current_question.strip()
    
    return a_dict, question, theme, sub_theme

 
start_page = 2
stop_page = 50 #pdfreader.numPages
text = ''

for x in range(start_page,stop_page):
    #create a variable that will select the selected number of pages
    pageobj=pdfreader.getPage(x+1)
    
    #(x+1) because python indentation starts with 0.
    #create text variable which will store all text datafrom pdf file
    text+=pageobj.extractText()

    
all_questions = list(map(lambda a:int(a.replace('-','').replace('\nQ','').strip()), re.findall("( \nQ\d+)",text)))
total_nr_of_questions = len(all_questions)
print(all_questions)

df = pd.DataFrame(np.zeros((total_nr_of_questions, 13)), columns = ['Q','Theme','Subtheme','-5','-4','-3','-2','-1','1','2','3','4','5'])

for i, question_nr in enumerate(all_questions):
    #print(question_nr)
    #question_nr = i
    try:
        answer_dict, question, theme, sub_theme = get_answer_dict(question_nr)
            ## Add question answers to dataframe
        for key, value in answer_dict.items():
            df.loc[i,'Q'] = question_nr
            df.loc[i,'Theme'] = theme
            df.loc[i,'Subtheme'] = sub_theme
            df.loc[i,key] = value
            
    except BaseException:
        print("Something wrong with question " + str(i))
        

       
    
df
```
